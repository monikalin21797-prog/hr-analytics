```python
#age
Q1 = hr_df['Age'].quantile(0.25)
Q3 = hr_df['Age'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = hr_df[(hr_df['Age'] < lower_bound) | (hr_df['Age'] > upper_bound)]

print("Lower bound:", lower_bound)
print("Upper bound:", upper_bound)
print("Number of outliers in Age:", outliers.shape[0])
print(outliers['Age'].unique())
```

```python
Q1 = hr_df['MonthlyIncome'].quantile(0.25)
Q3 = hr_df['MonthlyIncome'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = hr_df[(hr_df['MonthlyIncome'] < lower_bound) | (hr_df['MonthlyIncome'] > upper_bound)]

print("Lower bound:", lower_bound)
print("Upper bound:", upper_bound)
print("Number of outliers in MonthlyIncome:", outliers.shape[0])
print("Outlier values:", outliers['MonthlyIncome'].unique())

#Lower bound: -5291.0
#Upper bound: 16581.0
#Number of outliers in MonthlyIncome: 114



hr_df_no_outliers = hr_df[(hr_df['MonthlyIncome'] >= lower_bound) & (hr_df['MonthlyIncome'] <= upper_bound)]


```

BIVARIATE ANALYSIS

AGE VS ATTRITION

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Convert Attrition to numeric (Yes=1, No=0) for plotting
hr_df['Attrition_n'] = hr_df['Attrition'].map({'Yes': 1, 'No': 0})

# Group by Age and calculate Attrition rate
age_attrition = hr_df.groupby('Age')['Attrition_n'].mean().reset_index()

# Line plot
plt.figure(figsize=(10,6))
sns.lineplot(data=age_attrition, x='Age', y='Attrition_n', marker='o', color='blue')

plt.title("Attrition Rate vs Age")
plt.xlabel("Age")
plt.ylabel("Attrition Rate")
plt.show()
```

MONTHLY INCOME VS ATTRITION


```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
sns.boxplot(x='Attrition', y='MonthlyIncome', data=hr_df, palette='Set2')

plt.title("Attrition vs Monthly Income")
plt.xlabel("Attrition")
plt.ylabel("Monthly Income")
plt.show()
```

ATTRITION VS JOBPROFILE + SATISFACTION

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(14,6))
sns.barplot(data=hr_df,
            x="JobRole",
            y="JobSatisfaction",
            hue="Attrition",
            errorbar=None,
            palette="Set2")

plt.title("Attrition vs JobRole and JobSatisfaction")
plt.xlabel("Job Role")
plt.ylabel("Average Job Satisfaction")
plt.xticks(rotation=45)
plt.legend(title="Attrition")
plt.show()
```

attrition vr yearsatcompany + work life balance

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Convert Attrition to numeric (Yes=1, No=0)
hr_df['Attrition_n'] = hr_df['Attrition'].map({'Yes': 1, 'No': 0})

# Create pivot table: average attrition rate by YearsAtCompany & WorkLifeBalance
heatmap_data = hr_df.pivot_table(values='Attrition_n',
                                 index='YearsAtCompany',
                                 columns='WorkLifeBalance',
                                 aggfunc='mean')

plt.figure(figsize=(10,6))
sns.heatmap(heatmap_data, annot=True, cmap="YlGnBu", fmt=".2f")

plt.title("Attrition Rate: YearsAtCompany vs WorkLifeBalance")
plt.xlabel("Work Life Balance")
plt.ylabel("Years At Company")
plt.show()
```

SOLUTION TO BUSINESS OBJECTIVE

---



1) Reduce Employee Attrition
Target young and early-tenure employees with mentorship and onboarding programs to curb high turnover rates.
Conduct salary reviews to address lower income gaps observed in attrited staff.
Implement role-specific satisfaction surveys and action plans to boost job satisfaction.
Monitor work-life balance for new hires, offering flexible policies to reduce early exits.
2) Optimize Compensation Strategy
Benchmark salaries by role against industry standards, focusing on raising pay for lower earners.
Tie salary hikes and stock options to performance ratings for equitable rewards.
Analyze income variances across job levels to ensure competitive daily and hourly rates.
Introduce bonuses for overtime-heavy roles to align compensation with workload.
3) Improve Work-Life Balance & Employee Well-being
Mandate balance assessments in reviews, providing interventions for low ratings like reduced overtime.
Offer wellness programs and commute subsidies targeting high-risk groups with poor balance.
Promote flexible hours and remote options company-wide to prevent burnout in early years.
Track satisfaction metrics to refine policies, aiming to lower attrition in low-balance cohorts.
4) Enhance Career Development & Promotion Pathways
Flag stagnant employees (2+ years without promotion) for tailored training and role rotations.
Establish clear promotion milestones based on tenure, job level, and working years.
Prioritize development opportunities for at-risk age groups like under-25s and over-50s.
Increase annual training sessions to foster growth and reduce early-career stagnation.
5) Boost Employee Performance & Engagement
Launch role-targeted engagement initiatives like leadership training for low-satisfaction jobs.
Correlate involvement and ratings with satisfaction to design gamified learning programs.
Conduct team-building to enhance relationship satisfaction and prevent disengagement.
Use surveys to monitor and iterate on training, aiming for higher performance across roles.
6) Improve Workforce Planning & Talent Acquisition
Prioritize hiring mid-career profiles (ages 30-40) for high-attrition roles to match low-turnover patterns.
Refine recruitment by education field and travel preferences to align with retained employees.
Analyze department trends to optimize allocation and onboarding for better performance.
Simulate workforce scenarios using tenure and role data to predict and mitigate attrition risks.
